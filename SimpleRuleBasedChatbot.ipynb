```python
""" A simple rule-based chatbot that responds to user input based on predefined rules.
The chatbot can recognize greetings, farewells, and simple questions about the weather.
the rules are defined in a dictionary where the keys are keywords and the values are the corresponding responses.
the chatbot processes user input by checking for the presence of keywords and responding accordingly.
if no keywords are found, the chatbot responds with a default message."""
```

```python
class RuleBasedChatbot:
    def __init__(self):
        self.rules = {
            "hello": "Hello! How can I help you today?",
            "hi": "Hi there! What can I do for you?",
            "hey": "Hey! How's it going?",
            "bye": "Goodbye! Have a great day!",
            "goodbye": "See you later! Take care!",
            "weather": "The weather is sunny and warm today.",
            "how are you": "I'm just a chatbot, but I'm doing great! Thanks for asking."
        }
    
    def respond(self, user_input):
        user_input = user_input.lower()
        for keyword, response in self.rules.items():
            if keyword in user_input:
                return response
        return "I'm sorry, I don't understand. Can you please rephrase?"
    
if __name__ == "__main__":
    chatbot = RuleBasedChatbot()
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye! Have a great day!")
            break
        response = chatbot.respond(user_input)
        print(f"Chatbot: {response}")
```
